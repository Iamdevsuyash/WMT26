**#This Notebook converts the given dataset files(excel,csv) by WMT to jsnol files**

```python
import pandas as pd
import json
```

```python
# updates paths here
csv_file = "/content/drive/MyDrive/WMT_26_Machine_Translation/DataSet/mni-data/combined_en_mni.csv"

excel_file  = "/content/drive/MyDrive/WMT_26_Machine_Translation/DataSet/khasi-data/en_khasi.xlsx"

jsonl_file = "//content/drive/MyDrive/WMT_26_Machine_Translation/DataSet/khasi-data/en_kha.jsonl"

# df = pd.read_csv(csv_file)
df = pd.read_excel(excel_file)

with open(jsonl_file, "w", encoding="utf-8") as f:
    for _, row in df.iterrows():
        record = {
            "src": str(row.iloc[0]).strip(),
            "tgt": str(row.iloc[1]).strip(),
            "src_lang": "en_Latn",
            "tgt_lang": "mni_Beng"
        }
        f.write(json.dumps(record, ensure_ascii=False) + "\n")


print(f"Saved {jsonl_file}")
```

```python
with open(jsonl_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

for item in data:
    print(item)
    break
```
